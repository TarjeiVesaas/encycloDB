#Introducción al entorno de programación
El objetivo de esta práctica es el aprendizaje del entorno de programación Google Colab, que permite la escritura y ejecución de programas escritos en el lenguaje Python.
Un entorno de programación, también reconocido por su acrónimo en inglés IDE (Integrated Development Environment), es un programa que proporciona todas las herramientas que el programador necesita para desarrollar sus programas. Estas básicamente son:
* Un editor de texto para escribir el código fuente.
* Un compilador para obtener el código ejecutable.
* Un entorno para la ejecución segura de los programas creados.
* Un depurador para corregir los errores de ejecución.

En este caso, nosotros trabajamos con Python y con Google Colab y hay algunas de estas características que no tenemos disponibles como son el compilador (Python es un lenguaje interpretado por lo que no necesita ser compilado) y el depurador (para [depurar](https://colab.research.google.com/github/QuantEcon/lecture-python-programming.notebooks/blob/master/debugging.ipynb) con Google Colab hace falta utilizar una librería específica).

###Corrige los errores
Durante el ciclo de desarrollo de un programa pueden aparecer errores, los hay de dos tipos:

*   Errores de Ejecución: Se producen durante la ejecución del programa. Se detectan porque el funcionamiento del programa no es el deseado, debido a un fallo en la implementación o incluso en el diseño del programa (errores de diseño). Son los más difíciles de corregir y habitualmente hay que recurrir a la ayuda del IDE, al llamado depurador (o debbuger).
*   Errores de Compilación: Pueden deberse a fallos durante la escritura del programa (errores léxicos) o a faltas cometidas contra la sintaxis del lenguaje (errores sintácticos). Son los más fáciles de detectar y corregir, ya que el compilador los detecta y sugiere cuál puede ser la causa del error. Después de corregir los errores es necesario volver a compilar el programa. En nuestro caso, Python no utiliza compilador por lo que no tendremos este tipo de errores.
